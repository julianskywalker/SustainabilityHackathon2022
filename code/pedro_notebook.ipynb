```python
import numpy as np
import pandas as pd

import datetime
```

```python
input_fire_path = '../data/FW_Veg_Rem_Combined.csv'
input_city_path = '../data/city_attributes.csv'
input_humidity_path = '../data/humidity.csv'
input_pressure_path = '../data/pressure.csv'
input_tempetature_path = '../data/temperature.csv'
input_windD_path = '../data/wind_direction.csv'
input_windS_path = '../data/wind_speed.csv'
```

```python
df_fire = pd.read_csv(input_fire_path)
df_city = pd.read_csv(input_city_path)
df_hum = pd.read_csv(input_humidity_path)
df_pre = pd.read_csv(input_pressure_path)
df_tem = pd.read_csv(input_tempetature_path)
df_windD = pd.read_csv(input_windD_path)
df_windS = pd.read_csv(input_windS_path)
```

```python
df_dic = {'hum':df_hum, 'pre':df_pre, 'tem':df_tem, 'windD':df_windD, 'windS':df_windS}
```

```python
for key in df_dic.keys():
    df_dic[key]['datetime'] = pd.to_datetime(df_dic[key]['datetime']).apply(datetime.datetime.date)
    df_dic[key] = df_dic[key].groupby('datetime').mean()
df_hum, df_pre, df_tem, df_windD, df_windS = df_dic.values()
```
